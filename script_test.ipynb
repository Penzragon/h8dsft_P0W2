```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread('201030094143-stock-rhodesian-ridgeback-exlarge-169.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
Red = image[:,:,0]
Green = image[:,:,1]
Blue = image[:,:,2]

print('Shape R:',Red.shape)
print('Shape G:',Green.shape)
print('Shape B:',Blue.shape)
```

```python
fig, ax = plt.subplots(ncols=4,figsize=(21,5))
ax[0].imshow(image)
ax[0].set_title('Gambar Asli')
ax[1].imshow(Red)
ax[1].set_title('Gambar Filter R')
ax[2].imshow(Green)
ax[2].set_title('Gambar Filter G')
ax[3].imshow(Blue)
ax[3].set_title('Gambar Filter B')
```

```python
Grey = 0.299*Red + 0.587*Green + 0.114*Blue
```

```python
plt.imshow(Grey)
```

```python
Grey.shape
```

```python
dx, dy = np.gradient(Grey)
```

```python
fig, ax = plt.subplots(ncols=3,figsize=(16,5))
ax[0].imshow(Grey)
ax[0].set_title('Real Image')
ax[1].imshow(dx)
ax[1].set_title(r'$\partial_x img$')
ax[2].imshow(dy)
ax[2].set_title(r'$\partial_y img$')
```

```python
magnitude = np.sqrt(dx**2 + dy**2)
magnitude
```

```python
plt.imshow(magnitude)
```

```python
for i in range(magnitude.shape[0]-1):
    for j in range(magnitude.shape[1]-1):
        if magnitude[i][j] >= 25:
            magnitude[i][j] = 255
        else:
            magnitude[i][j] = 0
```

```python
plt.imshow(magnitude)
```
